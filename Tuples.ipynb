```python
# Creating tuples
t1 = (10, 20, 30, 40, 50)
t2 = (30, 40, 60)
print("t1:", t1)
print("t2:", t2)
```

```python
# Accessing elements
t1 = (10, 20, 30, 40, 50)
t2 = (30, 40, 60)
print("First element:", t1[0])
print("Last element:", t1[-1])
```

```python
# Slicing
t1 = (10, 20, 30, 40, 50)
t2 = (30, 40, 60)
print (t1[1:4])
```

```python
# Concatenation
t1 = (10, 20, 30, 40, 50)
t2 = (30, 40, 60)
t3 = t1 + t2
print("Concatenated tuple:", t3)
```
