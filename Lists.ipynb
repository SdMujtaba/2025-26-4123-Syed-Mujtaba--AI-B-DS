```python
# Creating lists
numbers = [10, 20, 30, 40]
print(numbers)
```

```python
# Access elements
numbers = [10, 20, 30, 40]
print(numbers[0])
print(numbers[-1])
```

```python
# Add element
numbers = [10,20,30,40]
numbers.append(50)
print(numbers)
```

```python
# Insert an element
numbers = [10,20,30,40]
numbers.insert(1, 15)
print(numbers)
```
