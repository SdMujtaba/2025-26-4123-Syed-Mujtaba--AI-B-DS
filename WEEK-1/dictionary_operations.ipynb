```python
# Creating a dictionary
d = {
    "name": "Mujtaba",
    "age": 19,
    "course": "BTech"
    }
print("Original dictionary:", d)
```

```python
d = {
    "name": "MJ",
    "age": 19,
    "course": "BTech"
}
# Accessing values
print("Name:", d["name"])
print("Age:", d.get("age"))
```

```python
d = {
    "name": "MJ",
    "age": 19,
    "course": "BTech"
}
# Adding a new key-value pair
d["college"] = "XYZ University"
print("After adding:", d)
```

```python
d = {
    "name": "MJ",
    "age": 19,
    "course": "BTech"
}
# Updating value
d["age"] = 21
print("After updating:", d)
```

```python
d = {
    "name": "MJ",
    "age": 19,
    "course": "BTech"
}
# Removing elements
d.pop("course")
print("After pop:", d)
removed_item = d.popitem()
print("After popitem:", d)
print("Removed item:", removed_item)
```

```python
d = {
    "name": "MJ",
    "age": 19,
    "course": "BTech"
}
# Keys, values, items
print("Keys:", d.keys())
print("Values:", d.values())
print("Items:", d.items())
```

```python
d = {
    "name": "MJ",
    "age": 19,
    "course": "BTech"
}
# Length
print("Length of dictionary:", len(d))
```

```python
d = {
    "name": "MJ",
    "age": 19,
    "course": "BTech"
}
# Looping through dictionary
print("Looping through keys:")
for key in d:
    print(key, ":", d[key])
print("Looping through key-value pairs:")
for k, v in d.items():
    print(k, "->", v)
```

```python

d = {
    "name": "MJ",
    "age": 19,
    "course": "BTech"
}
# Updating using update()
d.update({"city": "Hyderabad", "age": 22})
print("After update():", d)
```

```python

# Nested dictionary
student = {
    "id": 1,
    "info": {
        "name": "Amit",
        "marks": 90
    }
}
print("Nested dictionary value:", student["info"]["marks"])
```

```python
d = {
    "name": "MJ",
    "age": 19,
    "course": "BTech"
}
# Clearing dictionary
d.clear()
print("After clear:", d)
```
